#### Features
- id: only for test & sample submission files, id for prediction sample identification
- price: price in USD
- carat: weight of the diamond
- cut: quality of the cut (Fair, Good, Very Good, Premium, Ideal)
- color: diamond colour, from J (worst) to D (best)
- clarity: a measurement of how clear the diamond is (I1 (worst), SI2, SI1, VS2, VS1, VVS2, VVS1, IF (best))
- x: length in mm (longitud)
- y: width in mm (ancho)
- z: depth in mm (profundidad)
- depth: total depth percentage = z / mean(x, y) = 2 * z / (x + y) (43--79)
- table: width of top of diamond relative to widest point (43--95)

### Eliminar la colinealidad

```python
# Visualizo la correlación entre todo el dataFrame

corr = data.corr()

corr
```

```python
# Imprimo un mapa de calor para visualizar más fácilmente el grado de correlación

sns.heatmap(corr);
```

```python
# Fijándonos en la columna "price", las filas "carat","x","y" y "z" tienen mucha correlación con "price".
# De ellas vemos que entre "x","y" y "z" hay demasiada correlación, lo que nos indica que es probable que haya colinealidad
# Cómo "x" tiene mayor correlación con "price" decido dejarla y eliminar "y" y "z".

data = data.drop(["y","z"], axis="columns")
```

```python
# Separo los elementos categóricos de las columnas

data_dummy = pd.get_dummies(data, drop_first=True)

data_dummy.head()
```

```python
# Elimino y añado la columna "price" al final del dataFrame para poder compararlo mejor

data_dummy = data_dummy.drop(["price"], axis="columns")

data_dummy["price"] = y

data_dummy.head()
```

```python
# Obtengo las correlaciones de todo el dataFrame.

corr_dummy = data_dummy.corr()

corr_dummy
```

```python
# Vemos que ya ha desaparecido el problema de la colinealidad.

sns.heatmap(corr_dummy);
```

```python
# Disponemos el dataset para entrenarlo, por ello eliminamos la columna "price" de nuevo.

X = data_dummy.drop(["price"], axis="columns")

X.head()
```

## Selección del modelo y entrenamiento

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.20)
```

```python
from sklearn.linear_model import LinearRegression

# Elijo el modelo.

lr = LinearRegression()

lr
```

```python
# Entreno el modelo.

lr.fit = lr.fit(X_train, y_train)
```

```python
# Obtengo las predicciones tanto de X_test como de X_train, para compararlas entre ellas y ver si hay overfit

y_pred_test = lr.predict(X_test)
y_pred_train = lr.predict(X_train) 
```

```python
from sklearn.metrics import r2_score, mean_squared_error, mean_absolute_error

# Obtengo el error al cuadrado. Determina la calidad del modelo para replicar los resultado. 

r2_test = r2_score(y_test,y_pred_test)
r2_train = r2_score(y_train,y_pred_train)

print(f" El r2 de test es: {r2_test}")
print(f" El r2 de train es: {r2_train}")
```
