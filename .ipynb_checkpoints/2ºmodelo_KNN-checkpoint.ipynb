#### Features
- id: only for test & sample submission files, id for prediction sample identification
- price: price in USD
- carat: weight of the diamond
- cut: quality of the cut (Fair, Good, Very Good, Premium, Ideal)
- color: diamond colour, from J (worst) to D (best)
- clarity: a measurement of how clear the diamond is (I1 (worst), SI2, SI1, VS2, VS1, VVS2, VVS1, IF (best))
- x: length in mm (longitud)
- y: width in mm (ancho)
- z: depth in mm (profundidad)
- depth: total depth percentage = z / mean(x, y) = 2 * z / (x + y) (43--79)
- table: width of top of diamond relative to widest point (43--95)

### Separo los elementos categ√≥ricos de las columnas

```python
data_dummy = pd.get_dummies(data, drop_first=True)

data_dummy.head()
```

```python
# A√±ado la columna "y" para mirar c√≥mo correlaciona

data_dummy["price"] = y

data_dummy.head()
```

```python
corr = data_dummy.corr()

corr
```

```python
sns.heatmap(corr);
```

```python
y = data["price"]

y.head()
```

```python
X = data_dummy.drop(["price"], axis="columns")

X.head()
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.20)
```

```python
from sklearn.linear_model import LinearRegression

lr = LinearRegression()

lr
```

```python
lr.fit = lr.fit(X_train, y_train)
```

```python
y_pred_test = lr.predict(X_test)
y_pred_train = lr.predict(X_train) # Ver ovefit
```

```python
from sklearn.metrics import r2_score, mean_squared_error, mean_absolute_error

# Determina la calidad del modelo para replicar los resultado
# Primero el Ground truth (y), luego los valores objetivos estimados (y_pred)
# r2_score(y_true, y_pred)

r2_train = r2_score(y_train,y_pred_train)

r2_test = r2_score(y_test,y_pred_test)

print(r2_train)
print(r2_test)

if r2_train or r2_test > 0.6:
    print("La relaci√≥n es alta")
    
elif r2_train or r2_test < 0.4:
    print("La relaci√≥n es baja")
```

```python
from sklearn.metrics import mean_squared_error

mean_train = mean_squared_error(y_train,y_pred_train)**0.5

mean_test = mean_squared_error(y_test,y_pred_test)**0.5

print(mean_train)
print(mean_test)
```

```python
# cross_validation

from sklearn.model_selection import cross_val_score

neigh = lr

scoresKN = cross_val_score(neigh, X_train, y_train, cv=5, scoring="neg_root_mean_squared_error")

scoresKN
```
